```python
import pandas as pd
from analyticplots import time_bar_plot
import COVID19Py as cv
```

```python
covid = cv.COVID19()
br_location = covid.getLocationByCountryCode("BR", timelines=True)
br_timeline = br_location[0]["timelines"]["confirmed"]["timeline"]
data_br = pd.DataFrame().from_dict(data=br_timeline, orient="index")
data_br.columns = ["confirmed"]

time_bar_plot(data=data_br, y="confirmed", timestep="w", logy=True);
```

```python
data_br["country"] = "br"
ru_location = covid.getLocationByCountryCode("RU", timelines=True)
ru_timeline = ru_location[0]["timelines"]["confirmed"]["timeline"]
data_ru = pd.DataFrame().from_dict(data=ru_timeline, orient="index")
data_ru.columns = ["confirmed"]
data_ru["country"] = "ru"

data = data_br.append(data_ru)
time_bar_plot(data=data, y="confirmed", hue="country", timestep="M", norm=True, aggfunc="sum");
```
